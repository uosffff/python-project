```python

import pandas as pd

# 1. البيانات (d اختصار لـ Data)
d = {
    'id': ['A', 'B', 'C'], # نوع رأس الحفر
    'm': [100, 150, 120],  # الأمتار (meters)
    'h': [5, 10, 4]        # الساعات (hours)
}

# 2. تحويل لجدول (df اختصار لـ DataFrame)
df = pd.DataFrame(d)

# 3. حساب السرعة (r اختصار لـ ROP أو Rate)
df['r'] = df['m'] / df['h']

# 4. استخراج الصف الذي يحتوي على أعلى سرعة باستخدام loc
# idxmax تعطينا رقم الصف، و loc تجلب بياناته
top = df.loc[df['r'].idxmax()]

print(df)
print("\nالأفضل هو:")
print(top)

df.to_excel('report.xlsx', index=False)

print("تم إنشاء ملف الإكسيل بنجاح!")
```

```python

import pandas as pd

# 1. تجهيز بيانات تحتوي على تكرار لنفس رأس الحفر
d = {
    'id': ['A', 'A', 'B', 'B', 'C', 'C'],
    'm': [100, 150, 120, 130, 110, 140],  # الأمتار
    'h': [5, 7, 6, 6, 4, 8]              # الساعات
}
df = pd.DataFrame(d)

# 2. تجميع البيانات بناءً على النوع (id) وحساب المجموع
# هنا نستخدم groupby لتجميع الحروف المتشابهة ثم sum للمجموع
total_meters = df.groupby('id')['m'].sum()

# 3. حساب متوسط الساعات لكل نوع
avg_hours = df.groupby('id')['h'].mean()

print("--- إجمالي الأمتار لكل رأس حفر ---")
print(total_meters)
print("\n--- متوسط الساعات لكل رأس حفر ---")
print(avg_hours)
```
