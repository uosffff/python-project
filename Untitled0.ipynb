```python
import numpy as np
import matplotlib.pyplot as plt
days=["السبت","الاحد","الاثنين","الثلاثاء","الاربعاء" ]
sales=np.array([200,400,360,260,180])
print(f"toital:{sum(sales)}")
print(f"max:{max(sales)}")
colors=["royalblue","forestgreen","orange","crimson","purple"]

plt.bar(days,sales,color=colors)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Per Day")
plt.show()

```
