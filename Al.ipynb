```python


import streamlit as st
import pandas as pd
import sqlite3
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# ÿ•ÿπÿØÿßÿØ ŸÇÿßÿπÿØÿ© ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸÖÿ≠ŸÑŸäÿ© ŸÑÿ™ŸÑÿ®Ÿäÿ© ŸÖÿ™ÿ∑ŸÑÿ®ÿßÿ™ ÿßŸÑÿ≠ŸÅÿ∏ Offline
def init_db():
    conn = sqlite3.connect('edu_system.db')
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS students
                 (id INTEGER PRIMARY KEY, name TEXT, math INTEGER, science INTEGER,
                  engagement INTEGER, hours INTEGER, cluster INTEGER)''')
    conn.commit()
    conn.close()

# Ÿàÿ∏ŸäŸÅÿ© ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä (Preprocessing + Clustering)
def ai_logic(df):
    features = df[['math', 'science', 'engagement', 'hours']]
    scaler = StandardScaler()
    scaled = scaler.fit_transform(features)
    model = KMeans(n_clusters=3, random_state=42)
    df['cluster'] = model.fit_predict(scaled)
    return df

# Ÿàÿßÿ¨Ÿáÿ© ÿßŸÑŸÖÿ≥ÿ™ÿÆÿØŸÖ ÿßŸÑÿßÿ≠ÿ™ÿ±ÿßŸÅŸäÿ© (ÿßŸÑÿπÿ±ÿ®Ÿäÿ© ŸàÿßŸÑŸÖÿ™ÿ¨ÿßŸàÿ®ÿ©)
def main():
    st.set_page_config(page_title="ŸÜÿ∏ÿßŸÖ ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä ÿßŸÑÿ™ÿπŸÑŸäŸÖŸä", layout="wide")
    init_db()

    st.title("üìä ŸÜÿ∏ÿßŸÖ ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿ∑ŸÑÿßÿ® ÿßŸÑÿ∞ŸÉŸä")
    role = st.sidebar.selectbox("ÿßŸÑÿØŸàÿ± (RBAC)", ["ÿ∑ÿßŸÑÿ®", "ŸÖÿπŸÑŸÖ", "ŸÖÿ¥ÿ±ŸÅ", "ÿ¢ÿØŸÖŸÜ"])

    if role == "ŸÖÿπŸÑŸÖ":
        st.header("ŸÑŸàÿ≠ÿ© ÿßŸÑÿ™ÿ≠ŸÉŸÖ ŸÑŸÑŸÖÿπŸÑŸÖ")
        # ŸÜŸÖŸàÿ∞ÿ¨ ÿ•ÿØÿÆÿßŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
        with st.form("add_student"):
            name = st.text_input("ÿßÿ≥ŸÖ ÿßŸÑÿ∑ÿßŸÑÿ®")
            math = st.number_input("ÿßŸÑÿ±Ÿäÿßÿ∂Ÿäÿßÿ™", 0, 100)
            sci = st.number_input("ÿßŸÑÿπŸÑŸàŸÖ", 0, 100)
            eng = st.slider("ÿßŸÑÿ™ŸÅÿßÿπŸÑ", 1, 10)
            hrs = st.number_input("ÿßŸÑÿ≥ÿßÿπÿßÿ™ ÿßŸÑÿ£ÿ≥ÿ®ŸàÿπŸäÿ©", 1, 50)
            if st.form_submit_button("ÿ≠ŸÅÿ∏"):
                conn = sqlite3.connect('edu_system.db')
                conn.execute("INSERT INTO students (name, math, science, engagement, hours) VALUES (?,?,?,?,?)",
                             (name, math, sci, eng, hrs))
                conn.commit()
                st.success("ÿ™ŸÖ ÿßŸÑÿ≠ŸÅÿ∏!")

        # ÿ™ÿ¥ÿ∫ŸäŸÑ ÿßŸÑŸÄ AI
        if st.button("ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿ£ŸÜŸÖÿßÿ∑ Ÿàÿ™ŸàŸÑŸäÿØ ÿßŸÑÿÆÿ∑ÿ∑"):
            conn = sqlite3.connect('edu_system.db')
            df = pd.read_sql_query("SELECT * FROM students", conn)
            if len(df) >= 3:
                df = ai_logic(df)
                st.write(df)
                st.info("ÿ™ŸÖ ÿ™ŸàŸÑŸäÿØ ÿÆÿ∑ÿ∑ ÿßŸÑÿØÿ±Ÿàÿ≥ ÿ®ŸÜÿßÿ°Ÿã ÿπŸÑŸâ ÿßŸÑÿ™ÿµŸÜŸäŸÅ.")
            else:
                st.warning("ÿ£ÿØÿÆŸÑ ÿ®ŸäÿßŸÜÿßÿ™ 3 ÿ∑ŸÑÿßÿ® ÿπŸÑŸâ ÿßŸÑÿ£ŸÇŸÑ.")

if __name__ == "__main__":
    main()
```
